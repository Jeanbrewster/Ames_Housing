```python
plt.figure(figsize=(12, 9))
sns.heatmap(train.corr()[['SalePrice']].sort_values(by='SalePrice', ascending=False),
            cmap='coolwarm',
            vmax=1,
            vmin=-1,
            annot=True);
```

## Correlation summary
When analyzing the numerical columns, the variables with the strongest correlation(x>=60%) are overall quality (80%), general living area (70%), garage area(65%), number of cars held in the garage(65%), total basement sft(63%), first floor sft(62%). Correlations <60% have weaker relationships with price, year built(57%), year remodeled(55%), full bath(54%),garage year built(52%), Mas vnr area(51%), total rooms above ground(50%). 

```python
corr_matrix = train.corr()
corr_matrix['SalePrice'].sort_values(ascending=False)
```

```python
#showing a mostly normal distribution...
# add labels and takeaway
plt.figure(figsize=(12, 10))
plt.hist(train['Overall Qual']);
plt.title('Distribution of House Sales by Overall Quality'); 
plt.xlabel('House Quality');
plt.savefig('../Images/hist_Qual.jpeg')
```

```python
plt.figure(figsize=(12, 10))
plt.hist(train['SalePrice']);
plt.title('Distribution of Home Sales by Price'); 
plt.xlabel('Sale Price');
plt.savefig('../Images/hist_price.jpeg')
```

## Summary of Histograms
The histograms shows most of the numerical data is skewed. The sales price, or the outcome variable is skewed to the right and should be normalized for better predictive value. Overall quality is strongly correlated to sales price shows a mostly normal distribution.

```python
plt.figure(figsize=(12, 10))
stype_by_price= train.groupby(['House Style'])['SalePrice'].median().sort_values()
sns.boxplot(x=train['House Style'], y=train['SalePrice'], order=list(stype_by_price.index));
plt.title('House Style sorted by median sales price')
plt.savefig('../Images/box_HouseStyle.jpeg')
```

```python
plt.figure(figsize=(12, 10))
stype_by_price= train.groupby(['Sale Type'])['SalePrice'].median().sort_values()
sns.boxplot(x=train['Sale Type'], y=train['SalePrice'], order=list(stype_by_price.index));
plt.title('Sale type sorted by median sales price')
plt.savefig('../Images/box_SaleType.jpeg')
```

```python
# take a look at breakfast hour for visualization lesson
plt.figure(figsize=(12, 10))
sns.countplot(train['Neighborhood']);
plt.xticks(rotation=90)
plt.xlabel('Neighborhood', size=8)
plt.ylabel('Number of Home Sales')
plt.savefig('../Images/countplot_hood.jpeg')

```

```python
plt.figure(figsize=(12, 10))
hood_by_price= train.groupby(['Neighborhood'])['SalePrice'].median().sort_values()
sns.boxplot(x=train['Neighborhood'], y=train['SalePrice'], order=list(hood_by_price.index));
plt.title('Neighborhoods sorted by median sales price')
plt.xticks(rotation=90)
plt.savefig('../Images/box_hood.jpeg')
```

## Summary of Neighborhood Boxplots
Lower priced houses sold for a median price of around 100,000, while in more expensive neighborhoods, houses sold for around 300,000. In the lower-priced neighborhoods, the distribution of house prices is not as wide as in neighborhoods where houses were more expensive

```python
plt.figure(figsize=(12, 9))
plt.scatter(train['Overall Qual'], train['SalePrice'], color='purple', alpha=0.45)
plt.xlabel('Overall Qual')
plt.ylabel('SalePrice')
plt.title('Correlation of Sales Price to Overall Quality of the House')
plt.savefig('../Images/scatter_Qual.jpeg')
```

```python
plt.figure(figsize=(12, 9))
plt.scatter(train['Total Bsmt SF'], train['SalePrice'], color = 'navy')
#plt.plot(train['Total Bsmt SF'], train['SalePrice'], color = 'navy')
plt.xlabel('Basement Sq Ft')
plt.ylabel('SalePrice')
plt.title('Correlation of Sales Price to Total Basement Sq Ft')
plt.savefig('../Images/scatter_Basement.jpeg')
```

```python
plt.figure(figsize=(12, 9))
plt.scatter(train['Gr Liv Area'], train['SalePrice'], color = 'orange')
plt.xlabel('General Living Area')
plt.ylabel('SalePrice')
plt.title('Correlation of Sales Price to General Living Area')
plt.savefig('../Images/scatter_GrLivingArea.jpeg')
```
